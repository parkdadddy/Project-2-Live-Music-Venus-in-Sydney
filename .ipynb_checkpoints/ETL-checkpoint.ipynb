```python
import pandas as pd
```

```python
sydney_bar_df = pd.read_csv('templates/Raw Data/sydney_bars.csv')
sydney_bar_df
```

```python
perth_bar_df = pd.read_csv('templates/Raw Data/perth_bars.csv')
perth_bar_df
```

```python
frames = [perth_bar_df, sydney_bar_df]
bars_df = pd.concat(frames)

bars_df = bars_df.rename(columns={"geometry.location.lat": "Latitude",'geometry.location.lng':'Longitude'})
bars_cols = ["name",'vicinity','Latitude', 'Longitude',"name", "rating", "types",'business_status']
bars_df= bars_df[bars_cols].copy()
bars_df
```

```python
sydney_nc_df = pd.read_csv('templates/Raw Data/sydney_nc.csv')
sydney_nc_df
```

```python
perth_nc_df = pd.read_csv('templates/Raw Data/perth_nc.csv')
perth_nc_df
```

```python
nc_frames = [perth_nc_df, sydney_nc_df]
nc_df = pd.concat(nc_frames)

nc_df = nc_df.rename(columns={"geometry.location.lat": "Latitude",'geometry.location.lng':'Longitude'})
nc_cols = ["name",'vicinity','Latitude', 'Longitude',"rating", "types",'business_status']
nc_df= nc_df[nc_cols].copy()
nc_df
```

```python
sydney_bo_df = pd.read_csv('templates/Raw Data/sydney_bo.csv')
sydney_bo_df
```

```python
perth_bo_df = pd.read_csv('templates/Raw Data/perth_bo.csv')
perth_bo_df
```

```python
bo_frames = [perth_bo_df, sydney_bo_df]
bo_df = pd.concat(bo_frames)

bo_df = bo_df.rename(columns={"geometry.location.lat": "Latitude",'geometry.location.lng':'Longitude'})
bo_cols = ["name",'vicinity','Latitude', 'Longitude', "rating", "types",'business_status']
bo_df= bo_df[bo_cols].copy()
bo_df
```

```python
bo_df.to_csv('all_bottleo.csv')
nc_df.to_csv('all_nc.csv')
bars_df.to_csv('all_bars.csv')
```
